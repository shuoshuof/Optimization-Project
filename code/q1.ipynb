## Supplier Evaluation Metrics Formulas

The following metrics are used to evaluate supplier performance:

### 1. Average Supply Intensity

The average supply quantity across all time periods:

$$\text{ASI}_i = \frac{1}{T} \sum_{t=1}^{T} x_{i,t}$$

where:
- $\text{ASI}_i$ is the average supply intensity for supplier $i$
- $x_{i,t}$ is the supply quantity from supplier $i$ in period $t$
- $T$ is the total number of time periods

**Type**: Benefit indicator (larger is better)

---

### 2. Fulfillment Rate

The ratio of actual supply to order quantity (only for periods with orders):

$$\text{FR}_i = \frac{1}{|T_i|} \sum_{t \in T_i} \frac{x_{i,t}}{o_{i,t}}$$

where:
- $\text{FR}_i$ is the fulfillment rate for supplier $i$
- $o_{i,t}$ is the order quantity from supplier $i$ in period $t$
- $T_i = \{t : o_{i,t} > 0\}$ is the set of periods with orders
- $|T_i|$ is the number of periods with orders

**Type**: Interval indicator (optimal within range [0.8, 1.2])

---

### 3. Order Rate

The percentage of time periods with active supply (supply ≥ threshold):

$$\text{OR}_i = \frac{|\{t : x_{i,t} \geq \theta\}|}{T}$$

where:
- $\text{OR}_i$ is the order rate for supplier $i$
- $\theta$ is the threshold for considering a period as having active supply (default: 10)
- $|\{t : x_{i,t} \geq \theta\}|$ is the number of periods with supply ≥ threshold
- $T$ is the total number of time periods

**Type**: Benefit indicator (larger is better)

---

### 4. Supply Volatility

The coefficient of variation of supply quantities (for non-zero supply periods):

$$\text{SV}_i = \frac{\sigma_i}{\mu_i}$$

where:
- $\text{SV}_i$ is the supply volatility for supplier $i$
- $\sigma_i = \sqrt{\frac{1}{|T_i^+|} \sum_{t \in T_i^+} (x_{i,t} - \mu_i)^2}$ is the standard deviation of non-zero supply
- $\mu_i = \frac{1}{|T_i^+|} \sum_{t \in T_i^+} x_{i,t}$ is the mean of non-zero supply
- $T_i^+ = \{t : x_{i,t} > 0\}$ is the set of periods with positive supply
- $|T_i^+|$ is the number of periods with positive supply

**Type**: Cost indicator (smaller is better - lower volatility indicates more stable supply)

```python
class SupplierMetrics:
    """Supplier Metrics Calculation Class"""
    
    def __init__(self, data_order, data_supply):
        """
        Initialize
        
        Parameters:
        - data_order: Order data DataFrame
        - data_supply: Supply data DataFrame
        """
        self.data_order = data_order
        self.data_supply = data_supply
        self.week_cols = [col for col in data_order.columns if col.startswith('W')]
        # TODO: consider the type of material
        self.supplier_ids = data_order['供应商ID'].values
        self.n_suppliers = len(self.supplier_ids)
        
        # Dictionary to store all metrics
        self.metrics = {}
    
    def cal_metrics(self):
        """Calculate all predefined metrics"""
        print("Starting metrics calculation...")
        
        # 1. Average Supply Intensity
        self.metrics['Average Supply Intensity'] = self._cal_average_supply_intensity()
        
        # 2. Fulfillment Rate
        self.metrics['Fulfillment Rate'] = self._cal_fulfillment_rate()
        
        # 3. Order Rate
        self.metrics['Order Rate'] = self._cal_order_rate()

        # # 4. Risk Metric
        # self.metrics['Risk Metric'] = self._cal_risk()
        
        # 4. Supply Volatility
        self.metrics['Supply Volatility'] = self._cal_supply_volatility()

        print(f"Calculation completed! Total {len(self.metrics)} metrics calculated")
        return self

    def _cal_average_supply_intensity(self):
        """Calculate average supply intensity: average supply quantity across all weeks"""
        avg_supply_intensity = []
        for i in range(self.n_suppliers):
            supply = self.data_supply.iloc[i][self.week_cols].values
            avg_supply = np.mean(supply)
            avg_supply_intensity.append(avg_supply)
        
        return np.array(avg_supply_intensity)
    
    def _cal_fulfillment_rate(self):
        """Calculate fulfillment rate: supply / order ratio"""
        fulfillment_rate = []
        for i in range(self.n_suppliers):
            order = self.data_order.iloc[i][self.week_cols].values
            supply = self.data_supply.iloc[i][self.week_cols].values
            
            # Handle cases where order is 0
            mask = order > 0
            if mask.sum() > 0:
                rate = np.mean(supply[mask] / order[mask])
            else:
                rate = 0
            fulfillment_rate.append(rate)
        
        return np.array(fulfillment_rate)

    def _cal_order_rate(self):
        """Calculate order rate: percentage of weeks with orders >= threshold"""
        order_rate = []
        has_order_threshold = 10
        
        for i in range(self.n_suppliers):
            supply = self.data_supply.iloc[i][self.week_cols].values
            weeks_with_orders = np.sum(supply >= has_order_threshold)
            rate = (weeks_with_orders / len(self.week_cols))
            order_rate.append(rate)
        
        return np.array(order_rate)
    
    def _cal_risk(self):
        """Calculate risk metric (placeholder implementation)"""
        risk_metric = []
        for i in range(self.n_suppliers):
            supply = self.data_supply.iloc[i][self.week_cols].values
            risk = np.std(supply) / (np.mean(supply) + 1e-6)
            risk_metric.append(risk)
        
        return np.array(risk_metric)
    
    def _cal_supply_volatility(self):
        """Calculate supply volatility: coefficient of variation (std / mean)
        Only considers weeks with supply > 0 to avoid bias from zero values"""
        volatility = []
        
        for i in range(self.n_suppliers):
            supply = self.data_supply.iloc[i][self.week_cols].values
            # Only use non-zero supply weeks
            supply_nonzero = supply[supply > 0]
            
            if len(supply_nonzero) > 0 and np.mean(supply_nonzero) > 0:
                cv = np.std(supply_nonzero) / np.mean(supply_nonzero)
            else:
                cv = 0
            volatility.append(cv)
        
        return np.array(volatility)

    def get_metric_matrix(self):
        """
        Get metric matrix
        
        Returns:
        - DataFrame: rows are suppliers, columns are metrics
        """
        metric_df = pd.DataFrame(self.metrics, index=self.supplier_ids)
        metric_df.index.name = 'Supplier ID'
        return metric_df
    
    def add_custom_metric(self, metric_name, metric_values):
        """
        Add custom metric
        
        Parameters:
        - metric_name: Metric name
        - metric_values: Metric value array (length should match number of suppliers)
        """
        if len(metric_values) != self.n_suppliers:
            raise ValueError(f"Metric value count ({len(metric_values)}) does not match supplier count ({self.n_suppliers})")
        
        self.metrics[metric_name] = np.array(metric_values)
        print(f"Added custom metric: {metric_name}")
        
        return self
    
    def get_metric_summary(self):
        """Get metric statistical summary"""
        metric_df = self.get_metric_matrix()
        summary = metric_df.describe()
        return summary
```

## Entropy Weight Method

Calculate objective weights for each metric using entropy weight method

```python
class EntropyWeight:
    """Entropy Weight Method for objective weight calculation"""
    
    def __init__(self, data, indicator_types=None):
        """
        Initialize
        
        Parameters:
        - data: DataFrame with suppliers as rows and metrics as columns
        - indicator_types: Dict mapping metric names to types. Options:
            * 'benefit': larger is better (default for all metrics if not specified)
            * 'cost': smaller is better
            * 'interval': optimal within a range [best_min, best_max]
            * 'middle': optimal at a middle value
          Example: {'Supply Volatility': 'cost', 'Fulfillment Rate': 'benefit'}
        """
        self.data = data.copy()
        self.normalized_data = None
        self.weights = None
        self.entropy = None
        self.diversity = None
        
        # Set default indicator types (all benefit type)
        if indicator_types is None:
            self.indicator_types = {col: 'benefit' for col in data.columns}
        else:
            self.indicator_types = {}
            for col in data.columns:
                self.indicator_types[col] = indicator_types.get(col, 'benefit')
        
        # Store parameters for interval and middle types
        self.indicator_params = {}
    
    def set_indicator_params(self, metric_name, **params):
        """
        Set parameters for specific indicator types
        
        Parameters:
        - metric_name: Name of the metric
        - params: Parameters depending on indicator type
            * For 'interval' type: best_min, best_max
            * For 'middle' type: best_value
        
        Example:
        >>> model.set_indicator_params('Fulfillment Rate', best_min=0.95, best_max=1.05)
        >>> model.set_indicator_params('Supply Volatility', best_value=0.3)
        """
        self.indicator_params[metric_name] = params
        return self
    
    def normalize(self):
        """
        Preprocess and normalize indicators based on their types
        All indicators are converted to benefit type and normalized to [0, 1]
        """
        self.normalized_data = self.data.copy()
        
        for col in self.normalized_data.columns:
            values = self.normalized_data[col].values
            indicator_type = self.indicator_types[col]
            
            if indicator_type == 'benefit':
                # Benefit indicator: min-max normalization
                min_val = np.min(values)
                max_val = np.max(values)
                if max_val - min_val > 0:
                    self.normalized_data[col] = (values - min_val) / (max_val - min_val)
                else:
                    raise ValueError(f"All values are the same for benefit indicator '{col}'. Cannot normalize.")
                    
            elif indicator_type == 'cost':
                # Cost indicator: inverse then normalize
                min_val = np.min(values)
                max_val = np.max(values)
                if max_val - min_val > 0:
                    self.normalized_data[col] = (max_val - values) / (max_val - min_val)
                else:
                    raise ValueError(f"All values are the same for cost indicator '{col}'. Cannot normalize.")
                
            elif indicator_type == 'interval':
                # Interval indicator: optimal within [best_min, best_max]
                params = self.indicator_params.get(col, {})
                best_min = params.get('best_min', np.percentile(values, 25))
                best_max = params.get('best_max', np.percentile(values, 75))
                
                transformed = np.zeros_like(values, dtype=float)
                min_val = np.min(values)
                max_val = np.max(values)
                
                for i, val in enumerate(values):
                    if best_min <= val <= best_max:
                        transformed[i] = 1.0
                    elif val < best_min:
                        if max_val > best_min:
                            transformed[i] = 1 - (best_min - val) / (max_val - min_val)
                        else:
                            transformed[i] = 0.5
                    else:  # val > best_max
                        if max_val > best_min:
                            transformed[i] = 1 - (val - best_max) / (max_val - min_val)
                        else:
                            transformed[i] = 0.5
                
                self.normalized_data[col] = transformed
                
            elif indicator_type == 'middle':
                # Middle indicator: optimal at a specific value
                params = self.indicator_params.get(col, {})
                best_value = params.get('best_value', np.median(values))
                
                max_deviation = np.max(np.abs(values - best_value))
                if max_deviation > 0:
                    self.normalized_data[col] = 1 - np.abs(values - best_value) / max_deviation
                else:
                    self.normalized_data[col] = 1.0
        
        # Add small constant to avoid log(0) in entropy calculation
        self.normalized_data = self.normalized_data + 1e-10
        
        return self.normalized_data
    
    def calculate_weights(self):
        """Calculate entropy weights"""
        if self.normalized_data is None:
            self.normalize()
        
        n, m = self.normalized_data.shape  # n suppliers, m metrics
        
        # Calculate probability matrix
        P = self.normalized_data.div(self.normalized_data.sum(axis=0), axis=1)
        
        # Calculate entropy for each metric
        self.entropy = {}
        k = 1 / np.log(n)  # Constant k = 1/ln(n)
        
        for col in P.columns:
            p = P[col].values
            # Entropy: e_j = -k * sum(p_ij * ln(p_ij))
            entropy_j = -k * np.sum(p * np.log(p))
            self.entropy[col] = entropy_j
        
        # Calculate diversity coefficient (information utility value)
        self.diversity = {}
        for col in self.entropy.keys():
            # d_j = 1 - e_j
            self.diversity[col] = 1 - self.entropy[col]
        
        # Calculate weights
        diversity_sum = sum(self.diversity.values())
        self.weights = {}
        for col in self.diversity.keys():
            # w_j = d_j / sum(d_j)
            self.weights[col] = self.diversity[col] / diversity_sum
        
        return self.weights
    
    def get_weights_dataframe(self):
        """Get weights as DataFrame"""
        if self.weights is None:
            self.calculate_weights()
        
        weights_df = pd.DataFrame({
            'Metric': list(self.weights.keys()),
            'Type': [self.indicator_types[k] for k in self.weights.keys()],
            'Entropy': [self.entropy[k] for k in self.weights.keys()],
            'Diversity': [self.diversity[k] for k in self.weights.keys()],
            'Weight': list(self.weights.values())
        })
        
        return weights_df
    
    def calculate_weighted_score(self):
        """Calculate weighted comprehensive score for each supplier"""
        if self.weights is None:
            self.calculate_weights()
        
        # Use normalized data for scoring
        weighted_scores = np.zeros(len(self.normalized_data))
        
        for col in self.normalized_data.columns:
            weighted_scores += self.normalized_data[col].values * self.weights[col]
        
        return weighted_scores
```

## TOPSIS Method

Use TOPSIS (Technique for Order Preference by Similarity to Ideal Solution) to rank suppliers based on multiple criteria

```python
class TOPSIS:
    """TOPSIS Method for multi-criteria decision making
    
    Note: Assumes all indicators have been preprocessed to benefit type 
    (larger is better) by entropy weight normalization.
    """
    
    def __init__(self, data, weights):
        """
        Initialize TOPSIS
        
        Parameters:
        - data: DataFrame with decision units as rows and criteria as columns
                All indicators should already be normalized to benefit type
        - weights: Dict or array of weights for each criterion (must sum to 1)
        """
        self.data = data.copy()
        self.normalized_data = None
        self.weighted_normalized_data = None
        self.ideal_solution = None
        self.negative_ideal_solution = None
        self.distances_to_ideal = None
        self.distances_to_negative_ideal = None
        self.scores = None
        
        # Set weights
        if isinstance(weights, dict):
            self.weights = np.array([weights[col] for col in data.columns])
        else:
            self.weights = np.array(weights)
        
        # Verify weights sum to 1
        if not np.isclose(self.weights.sum(), 1.0):
            raise ValueError(f"Weights must sum to 1, got {self.weights.sum()}")
    
    def normalize(self):
        """
        Normalize the decision matrix using vector normalization
        x_ij' = x_ij / sqrt(sum(x_ij^2))
        """
        self.normalized_data = self.data.copy()
        
        for col in self.data.columns:
            values = self.data[col].values
            norm = np.sqrt(np.sum(values ** 2))
            if norm > 0:
                self.normalized_data[col] = values / norm
            else:
                self.normalized_data[col] = 0
        
        return self.normalized_data
    
    def apply_weights(self):
        """Apply weights to normalized matrix"""
        if self.normalized_data is None:
            self.normalize()
        
        self.weighted_normalized_data = self.normalized_data.copy()
        for i, col in enumerate(self.normalized_data.columns):
            self.weighted_normalized_data[col] = self.normalized_data[col] * self.weights[i]
        
        return self.weighted_normalized_data
    
    def determine_ideal_solutions(self):
        """
        Determine ideal and negative ideal solutions
        Since all indicators are already benefit type:
        - Ideal solution: maximum values
        - Negative ideal solution: minimum values
        """
        if self.weighted_normalized_data is None:
            self.apply_weights()
        
        self.ideal_solution = {}
        self.negative_ideal_solution = {}
        
        for col in self.weighted_normalized_data.columns:
            values = self.weighted_normalized_data[col].values
            self.ideal_solution[col] = np.max(values)
            self.negative_ideal_solution[col] = np.min(values)
        
        return self.ideal_solution, self.negative_ideal_solution
    
    def calculate_distances(self):
        """Calculate Euclidean distances to ideal and negative ideal solutions"""
        if self.ideal_solution is None:
            self.determine_ideal_solutions()
        
        n = len(self.weighted_normalized_data)
        self.distances_to_ideal = np.zeros(n)
        self.distances_to_negative_ideal = np.zeros(n)
        
        for i in range(n):
            sum_ideal = 0
            sum_negative_ideal = 0
            
            for col in self.weighted_normalized_data.columns:
                value = self.weighted_normalized_data.iloc[i][col]
                sum_ideal += (value - self.ideal_solution[col]) ** 2
                sum_negative_ideal += (value - self.negative_ideal_solution[col]) ** 2
            
            self.distances_to_ideal[i] = np.sqrt(sum_ideal)
            self.distances_to_negative_ideal[i] = np.sqrt(sum_negative_ideal)
        
        return self.distances_to_ideal, self.distances_to_negative_ideal
    
    def calculate_scores(self):
        """Calculate TOPSIS scores (relative closeness to ideal solution)
        Score = D- / (D+ + D-)
        where D+ is distance to ideal, D- is distance to negative ideal
        """
        if self.distances_to_ideal is None:
            self.calculate_distances()
        
        # Avoid division by zero
        denominator = self.distances_to_ideal + self.distances_to_negative_ideal
        self.scores = np.where(
            denominator > 0,
            self.distances_to_negative_ideal / denominator,
            0.5
        )
        
        return self.scores
    
    def get_ranking(self):
        """Get ranking results with scores"""
        if self.scores is None:
            self.calculate_scores()
        
        result_df = self.data.copy()
        result_df['TOPSIS Score'] = self.scores
        result_df['Rank'] = result_df['TOPSIS Score'].rank(ascending=False, method='min').astype(int)
        # Sort by rank
        # Sort by rank
        result_df = result_df.sort_values('Rank')
        
        return result_df
        
    def get_summary(self):
        """Get summary information"""
        if self.scores is None:
            self.calculate_scores()
        
        summary = {
            'Ideal Solution': self.ideal_solution,
            'Negative Ideal Solution': self.negative_ideal_solution,
            'Score Statistics': {
                'Mean': self.scores.mean(),
                'Std': self.scores.std(),
                'Min': self.scores.min(),
                'Max': self.scores.max()
            }
        }    
        return summary  
```
